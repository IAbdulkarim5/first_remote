```python
average_ages = df.groupby('Survived')['Age'].mean().reset_index()
sns.barplot(x='Survived', y='Age', data=average_ages)

plt.show()

## نسبة المتوفين اعلى من اللذين نجو
```

```python

death_survived = df.groupby(['Sex', 'Survived']).size().reset_index(name='Count')

plt.figure(figsize=(10, 6))

sns.barplot(x='Sex', y='Count', hue='Survived', data=death_survived)

plt.title('Gender_Alive', fontsize=16)
plt.xlabel('Sex', fontsize=12)
plt.ylabel('numbers', fontsize=12)
plt.ylim(0, death_survived['Count'].max() + 10)  


plt.show()

## نسبة وفاة الرجال اكثر من النساء 
```
